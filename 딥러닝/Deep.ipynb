```python
import numpy as np
import matplotlib.pylab as plt



 
def sigmoid( x):

     return 1 / (1 + np.exp(-x))

 

# 그래프 그리기
X = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(X)
plt.plot(X, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# Tanh 함수 (하이퍼볼릭 탄젠트 함수)

import numpy as np
import matplotlib.pylab as plt





def tanh(x):

    return np.tanh(x)


# 그래프 그리기
x = np.arange(-5.0, 5.0, 0.1)
y = tanh(x)
plt.plot(x, y)
plt.ylim(-1.1, 1.1)
plt.title("탄젠트 함수")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pylab as plt




def relu(x):

     return np.maximum(0, x)

 

# 그래프 그리기
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-0.5, 5.5)
plt.title("reLU activation function")
plt.grid()
plt.show()
```

```python
# keras를 활용한 y = 2x + 1 선형회귀 모델 예측 예제
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#   1. 데이터 준비
#   입력(x)과 출력(y) 데이터를 정의합니다.
x_data = np.array([0, 1, 2, 3, 4], dtype=np.float32)    # 입력
y_data = np.array([1, 3, 5, 7, 9], dtype=np.float32)    # 정답

# 2. 모델 정의
# Sequential API를 사용해 신경망을 정의합니다.
model = Sequential([
    Dense(1, input_dim=1)  # 입력 1개, 출력 1개 (선형 회귀)
])

# 3. 모델 컴파일
# 손실 함수와 옵티마이저를 설정합니다.
model.compile(optimizer='sgd', loss='mse')


# 4. 모델 학습
# fit() 메서드로 데이터를 학습시킵니다.
model.fit(x_data, y_data, epochs=50, verbose=1)  # 50번 반복 학습
 
# 5. 예측 및 평가
# 새로운 데이터로 예측 수행
x_new = np.array([5], dtype=np.float32)
y_new = model.predict(x_new)

print(f"x = 5일때, 예측된 y = {y_new[0][0]:.2f}")
```

```python
# 꽃 이미지 분류
# flower_photo 디렉토리 밑에 daisy, dandelion, roses, sunflowers, tulips 디렉토리 존재, 약 3670개 사진
import matplotlib.pyplot as plt
import numpy as np
import PIL
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.models import load_model

# 운영체제
import os
import pathlib
import urllib.request
import tarfile


# 데이터 다운로드 함수
def download_and_extract_data():

    # URL
    # dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"

    # # 디렉토리
    # download_dir = os.path.join(os.getcwd(), 'flower_dataset')
    # os.makedirs(download_dir, exist_ok=True)

    # # 다운로드
    # tgz_path = os.path.join(download_dir, "flower_photos.tgz")

    # if not os.path.exists(tgz_path):
    #     print("데이터셋 다운로드 중")
    #     urllib.request.urlretrieve(dataset_url, tgz_path)
    #     print("다운로드 완료")

    # # 압축 해제
    # if not os.path.exists(os.path.join(download_dir, "flower_photos")):
    #     print("압축 해제 중")
    #     with tarfile.open(tgz_path) as tar:
    #         tar.extractall(path=download_dir)
    #     print("압축 해제 완료")

    #     return os.path.join(download_dir, "flower_photos")

    # 데이터 URL
    dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"

    # 다운로드할 디렉토리 생성
    # 현재 디렉토리, 폴더이름
    download_dir = os.path.join(os.getcwd(), "flower_dataset")
    # A에 B를 합치기
    os.makedirs(download_dir, exist_ok=True)
    # 폴더 생성

    # 파일 다운로드
    tgz_path = os.path.join(download_dir, "flower_photos.tgz")
    # 폴더에 파일을 넣기
    # 폴더에 압축파일이 없다면
    if not os.path.exists(tgz_path):
        print("데이터셋 다운로드 중...")
        urllib.request.urlretrieve(dataset_url, tgz_path)
        print("다운로드 완료!")

    # 압축 해제
    # dataset 폴더에 photos폴더가 없다면
    if not os.path.exists(os.path.join(download_dir, "flower_photos")):
        print("압축 해제 중...")
        with tarfile.open(tgz_path) as tar:
            tar.extractall(path=download_dir)
        print("압축 해제 완료!")

    # photo 폴더 돌려주기
    return os.path.join(download_dir, "flower_photos")



# 2. 데이터 다운도르 및 경로 설정
data_dir = download_and_extract_data()
data_dir = pathlib.Path(data_dir)

# *클래스 및 이미지 개수 확인
classes = sorted([item.name for item in data_dir.glob("*/") if item.is_dir()])
image_count = len(list(data_dir.glob("*/*.jpg")))
print(f"발견된 클래스: {classes}")
print(f"총 이미지 개수: {image_count}")

# *클래스 별 이미지 개수 확인
for image in classes:
    class_image = len(list(data_dir.glob(f"{image}/*")))
    print(f"{image}: {class_image}개 이미지")




# 이미지 설정
img_height = 180
img_width = 180
batch_size = 200

# 훈련 데이터 셋 생성 (데이터 로드)
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    str(data_dir),
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size,
)

# 검증 데이터 셋 설정
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    str(data_dir),
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size,
)

# *클래스 이름 확인
class_name = train_ds.class_names
print("클래스:", class_name)
num_class = len(class_name)

# 데이터 셋 "성능 최적화" (AUTOTUNE)
autotune = tf.data.AUTOTUNE
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=autotune)
val_ds = val_ds.cache().prefetch(buffer_size=autotune)

# 데이터 증강 레이어 정의
data_augmentation = tf.keras.Sequential(
    [
        layers.RandomFlip("horizontal"),
        layers.RandomRotation(0.2),
        layers.RandomZoom(0.2),
    ]
)

# *<샘플 이미지> 시각화
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_name[labels[i]])
        plt.axis("off")
plt.show()

# CNN 모델 구성. 생성 (중요)
model = tf.keras.Sequential(
    [
        # 데이터 증강 레이어 정의
        data_augmentation,
        # 정규화
        layers.Rescaling(1.0 / 255, input_shape=(img_height, img_width, 3)),
        # 필터 #3X3
        layers.Conv2D(16, 3, padding="same", activation="relu"),
        layers.MaxPool2D(),
        layers.Conv2D(32, 3, padding="same", activation="relu"),
        layers.MaxPool2D(),
        layers.Conv2D(64, 3, padding="same", activation="relu"),
        layers.MaxPool2D(),
        layers.Dropout(0.2),
        layers.Flatten(),
        layers.Dense(128, activation="relu"),
        layers.Dense(num_class, activation="softmax"),
    ]
)

# 모델 컴파일 (모델 직접만들기)
model.compile(
    # 최적화
    optimizer="adam",
    # 손실률
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),  # 다중 클래스서 사용, 각 클래스별 확률 예측 및 손실 계산
    # 평가지표
    metrics=["accuracy"],
)

# *모델 요약(구조 출력. head)
model.summary()

# 얼리 스탑 콜백 설정
early_stop = [
    
    tf.keras.callbacks.EarlyStopping(
        monitor="val_loss", patience=3, restore_best_weights=True
    )
    
]

# 모델 학습 (fit)
epochs = 15
history = model.fit(
                                      # 반복          # "얼리스탑"
    train_ds, validation_data=val_ds, epochs=epochs, callbacks=early_stop
)


# *<학습 결과> 시각화
acc = history.history["accuracy"]
val_acc = history.history["val_accuracy"]
loss = history.history["loss"]
val_loss = history.history["val_loss"]

epochs_range = range(len(acc))

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label="Training Accuracy")
plt.plot(epochs_range, val_acc, label="Validation Accuracy")
plt.title("Training and Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(loc="lower right")

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label="Training Loss")
plt.plot(epochs_range, val_loss, label="Validation Loss")
plt.title("Training and Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(loc="upper right")
plt.show()

# 모델 저장
# 폴더 생성 (save)
model.save("./model/flower_classification_model.h5")
print("모델이 저장되었습니다")

# 저장 모델 불러오기
model_path = "./model/flower_classification_model.h5"
loaded_model = load_model(model_path)
print("저장된 모델이 성공적으로 로드되었습니다")
#


# !저장된 모델을 활용한 "이미지" 예측 함수
def predict_flower_with_loaded_model(img_path, model, class_names):
    """
    이미지 경로와 모델을 입력받아 예측 결과를 출력하고 이미지를 시각화하는 함수.

    Parameters:
        img_path (str): 예측할 이미지 파일 경로.
        model (tf.keras.Model): 저장된 모델.
        class_names (list): 클래스 이름 리스트.
    """

    # 이미지 로드 및 전처리
    img = tf.keras.preprocessing.image.load_img(
        img_path, target_size=(img_height, img_width)
    )
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)

    # 예측
    predictions = model.predict(img_array)
    score = tf.nn.softmax(predictions[0])

    # 예측 결과 출력
    print(
        "이 이미지는 '{}' ({:.2f}% 확률) 입니다.".format(
            class_names[np.argmax(score)], 100 * np.max(score)
        )
    )

    # <함수> 이미지 시각화
    plt.imshow(img)
    # 축의 범위 끄기
    # axis.off > 이미지만 보여줌
    # axis 없음 > 설정에 따라 눈금, 축 등이 생김
    # plt.axis('off')
    plt.show()


# 테스트 이미지 경로
test_image_path = "./flower_dataset/flower_photos/roses/12240303_80d87f77a3_n.jpg"

# 예측 실행
predict_flower_with_loaded_model(test_image_path, loaded_model, class_name)
```
